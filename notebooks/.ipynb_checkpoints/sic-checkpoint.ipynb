```python
import xarray as xr
import numpy as np
import matplotlib.ticker as ticker
from matplotlib import pyplot as plt
import matplotlib.path as mpath
from cartopy.util import add_cyclic_point
import cartopy.crs as ccrs
```

```python
dirpath = '/Users/graemem/Documents/research/data/sic_nsidc_cdr/'
grid = xr.open_dataset(dirpath+'gridfile_sic.nc')

# sic = xr.open_dataset(dirpath + 'v04r00/seaice_conc_monthly_sh_201401-202012_v04r00_merged.nc')
sic = xr.open_dataset(dirpath + 'v04r00/seaice_conc_monthly_clim_sh_201401-202012_v04r00_merged.nc')

ds = xr.merge([grid,sic])
```

```python
month = '09'
dsnow = ds.sel({'time':'2020-'+month+'-01'})
```

```python
Z = dsnow['cdr_seaice_conc_monthly'].copy()
X = dsnow['lon'].copy()
Y = dsnow['lat'].copy()

extent = (lon1, lon2, lat1, lat2) = (-210, -130, -90, -50)
rect = mpath.Path([[lon1, lat1], [lon2, lat1],
    [lon2, lat2], [lon1, lat2], [lon1, lat1]]).interpolated(50)

clon = 0.5*(lon1+lon2)
clat = 0.5*(lat1+lat2)
crs_plt = ccrs.EquidistantConic(central_longitude=clon,
                                central_latitude=clat,
                                standard_parallels=clat)
crs_src = ccrs.PlateCarree()

fig,ax=plt.subplots(figsize=(10,10),subplot_kw={'projection':crs_plt})
ax.coastlines()
ax.set_extent(extent,crs_src)
ax.set_boundary(rect,crs_src)
ax.plot(np.array([-160,-150,-140]),np.array([-70,-70,-70]),transform=crs_src)
```
